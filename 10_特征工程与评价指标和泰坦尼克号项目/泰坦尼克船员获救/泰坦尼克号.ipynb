````python
#### 处理缺失数据# 有3个特征头缺失值
# Age, Cabin,  Embarkerd
# Age:年龄
#   714个非空，177个空，填充
#    我们认为年龄对存活率有一定影响，考虑保留改特征，填充

# Cabin :门牌号
#   204个非空 ，787个空，空的比较多，可以考虑删除该特征

# Embarkerd：从哪里上船
#    889个非空，2个空，填充
````

````python
titanic.Embarked.value_counts() #S    646, C    168, Q     77

# 从哪里上船 Embarked
titanic.Embarked = titanic.Embarked.map({"S": 0, "C": 1, "Q": 2})
````

### 使用线性回归预测, 知识点:交叉验证
- 这里是分类问题， 我们尝试使用普通线性回归
- from sklearn.linear_model import LinearRegression
- from sklearn.model_selection import StratifiedKFold

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import StratifiedKFold
````

````python
#  Stratified K-Folds cross-validator（分成k折交叉验证器）
skf = StratifiedKFold() #默认是5折
````

````python
list(skf.split(data, target))
````

````python
# 平均得分
scores = []
# 遍历5次
for train, test in skf.split(data, target):
#     print(train, test)
    
    # 训练数据，测试数据
    X_train = data.iloc[train]
    y_train = target.iloc[train]
    X_test = data.iloc[test]
    y_test = target.iloc[test]
    
    # 线性回归
    linear = LinearRegression()
    linear.fit(X_train,y_train)
    
    # 预测
    y = linear.predict(X_test)
    # print(y)
    
    # 自己设定一个阈值，把回归问题转换成分类问题（类似逻辑斯蒂回归）
    # >=0.5 认为是正例子，反之则为负例 (得救or没得救)
    y_predict = y >0.6
    # print(y_predict)
    
    # 得分： 准确率
    # 如果预测对了就返回 1，错了就返回 0
    y_score = (y_predict == y_test)
    
    # 计算得分，预测对了的 在测试样本中的占比
    # 取平均值
    print(y_score.mean())
    
    scores.append(y_score.mean())

np.array(scores).mean()
````

##### 增加`FamilySize`家庭成员个数 , `NameLength`名字长度, `Title`头衔

````python
titanic.head()
````

````python
# 特征衍生： 利用现有特征进行组合， 变换等方式生成新的特征

# 增加FamilySize家庭成员个数
titanic['FamilySize'] = titanic.SibSp + titanic.Parch
````

````python
# 增加特征NameLength名字长度
titanic['NameLength'] = titanic['Name'].map(lambda x: len(x))
````

````python
# 增加特征头衔Title
titanic['Title'] = titanic['Name'].map(lambda x: x.split('.')[0].split(', ')[1])
titanic['Title'] = titanic['Title'].map({"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Dr": 5, "Rev": 6, "Major": 7, "Col": 7, "Mlle": 8, "Mme": 8, "Don": 9, "Lady": 10, "the Countess": 10, "Jonkheer": 10, "Sir": 9, "Capt": 7, "Ms": 2})
````
